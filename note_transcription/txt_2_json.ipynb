```python
from mir_eval.util import hz_to_midi
import csv
import numpy as np
from mutagen.mp3 import MP3



def load_midi(file_ID, level):
    
    audio = MP3(level + file_ID + '.mp3')
    total_duration = audio.info.length

    notes_MIDI = []
    with open(level + file_ID + '.txt', 'rb') as csvfile:
                score_ = csv.reader(csvfile, delimiter='\t')
                for row in score_:
                        start_ts =  float(row[0]) / total_duration
                        end_ts =  (float(row[0]) + float(row[2]) ) / total_duration
                        midi_freq = hz_to_midi( float(row[1]) )
                        notes_MIDI.append([start_ts, end_ts, midi_freq])
    return notes_MIDI
```

```python

track_ID = 'Rihanna'
level = 'One_Line_Regions/'
level = 'Smaller/'

total_max = 0
total_min = 1000
puzzle_pieces = []
for i in range(1,9):
    a = load_midi(track_ID + str(i), level)
    b = np.array(a)
        
    puzzle_pieces.append(a)
    
    max_note_MIDI = np.max(b[:,2])
    min_note_MIDI = np.min(b[:,2])
    if max_note_MIDI > total_max:
        total_max = max_note_MIDI
    if min_note_MIDI < total_min:
        total_min = min_note_MIDI

print total_min
print total_max
```
